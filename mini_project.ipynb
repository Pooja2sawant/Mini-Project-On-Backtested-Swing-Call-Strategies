```python
# --- Import Packages ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import mplfinance as mpf
import feedparser
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# --- Load the Data ---
tickerData = yf.Ticker('RELIANCE.NS')
tickerDf = tickerData.history(period='600d', interval='1d')
df = tickerDf.drop(['Dividends', 'Stock Splits'], axis=1)
df = df.round(2)
print(df.head())

# --- Visualize the Stock ---
mpf.plot(df.iloc[-100:, :], type='candle', style='charles', volume=True, figscale=1.2)

# --- Create Moving Averages ---
df['20EMA'] = df['Close'].ewm(span=20, adjust=False).mean().round(2)
df['50EMA'] = df['Close'].ewm(span=50, adjust=False).mean().round(2)
print(df.tail(10))

# --- Visualize the Stock with EMA ---
EMA = [
    mpf.make_addplot(df['20EMA'], color='blue'),
    mpf.make_addplot(df['50EMA'], color='black')
]
mpf.plot(df, addplot=EMA, type='candle', style='charles', volume=False, figscale=1.1)

# --- Create a Function to Buy and Sell ---
def trade_call(df):
    Buy_price = []
    Sell_price = []
    signal = 0

    for i in range(len(df)):
        if df['20EMA'][i] > df['50EMA'][i]:
            if signal != 1:
                Buy_price.append(df['Close'][i])
                Sell_price.append(np.nan)
                signal = 1
            else:
                Buy_price.append(np.nan)
                Sell_price.append(np.nan)
        elif df['20EMA'][i] < df['50EMA'][i]:
            if signal != -1:
                Buy_price.append(np.nan)
                Sell_price.append(df['Close'][i])
                signal = -1
            else:
                Buy_price.append(np.nan)
                Sell_price.append(np.nan)
        else:
            Buy_price.append(np.nan)
            Sell_price.append(np.nan)

    return Buy_price, Sell_price

df['Buy_Signal'], df['Sell_Signal'] = trade_call(df)

# --- Visualize Stock with Signals ---
df_plot = df.tail(200)
addplots = [
    mpf.make_addplot(df_plot['20EMA'], color='blue'),
    mpf.make_addplot(df_plot['50EMA'], color='black'),
    mpf.make_addplot(df_plot['Buy_Signal'], type='scatter', markersize=100, marker='^', color='green'),
    mpf.make_addplot(df_plot['Sell_Signal'], type='scatter', markersize=100, marker='v', color='red')
]
mpf.plot(df_plot, addplot=addplots, type='candle', style='charles', volume=True, figscale=1.2)

# --- Create Portfolio ---
def portfolio(df):
    df1 = pd.DataFrame()
    df1['Buy_Price'] = df['Buy_Signal']
    df1['Sell_Price'] = df['Sell_Signal']
    df1 = df1.reset_index(drop=True)
    first_entry = df1['Buy_Price'].first_valid_index()
    last_exit = df1['Sell_Price'].last_valid_index()
    df1 = df1.loc[first_entry:last_exit]
    df1 = df1.apply(lambda x: pd.Series(x.dropna().values))
    df1['Returns'] = df1['Sell_Price'] - df1['Buy_Price']
    df1['Percentage Returns'] = ((df1['Returns'] / df1['Buy_Price']) * 100).round(2).astype(str) + '%'
    df1['Profit or Loss'] = df1['Returns'].apply(lambda x: 1 if x > 0 else -1)
    return df1

df1 = portfolio(df)
print(df1)

# --- Step 1: Fetch News Headlines from Google News ---
def get_news_headlines():
    query = "Reliance Industries"
    url = f"https://news.google.com/rss/search?q={query.replace(' ', '+')}&hl=en-IN&gl=IN&ceid=IN:en"
    feed = feedparser.parse(url)
    headlines = [entry.title for entry in feed.entries]
    return headlines

# --- Step 2: Perform Sentiment Analysis ---
def analyze_sentiment(news_list):
    sia = SentimentIntensityAnalyzer()
    sentiment = {'positive': 0, 'neutral': 0, 'negative': 0}

    for line in news_list:
        score = sia.polarity_scores(line)
        if score['compound'] >= 0.05:
            sentiment['positive'] += 1
        elif score['compound'] <= -0.05:
            sentiment['negative'] += 1
        else:
            sentiment['neutral'] += 1
    return sentiment

# --- Step 3: Run Analysis ---
news_data = get_news_headlines()
sentiment_scores = analyze_sentiment(news_data)

# --- Step 4: Print Sentiment Summary ---
print(f"\nSentiment Analysis for Reliance Industries:")
print(f"Total Headlines: {len(news_data)}")
print(f"Positive : {sentiment_scores['positive']}")
print(f"Neutral  : {sentiment_scores['neutral']}")
print(f"Negative : {sentiment_scores['negative']}")

# --- Step 5: Plot Pie Chart ---
if len(news_data) > 0:
    labels = ['Positive', 'Neutral', 'Negative']
    sizes = [sentiment_scores['positive'], sentiment_scores['neutral'], sentiment_scores['negative']]
    colors = ['green', 'gold', 'red']

    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    plt.title('Sentiment Distribution for Reliance Industries News')
    plt.axis('equal')
    plt.show()
else:
    print("\nNo news headlines were found. Pie chart will not be displayed.")

# --- Educational Building Blocks ---
def show_educational_building_blocks():
    print("\nðŸ“š Educational Building Blocks:")
    blocks = [
        "What is a Stock?",
        "Understanding Candlestick Charts",
        "What is a Moving Average?",
        "Difference Between Technical & Fundamental Analysis"
    ]
    for b in blocks:
        print(f"- {b}")
show_educational_building_blocks()

# --- Basic Market Concepts with Visuals ---
def visualize_basic_concepts(df):
    plt.figure(figsize=(10, 5))
    plt.plot(df['Close'], label='Close Price')
    plt.title("Basic Concept: Stock Closing Price Trend")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid()
    plt.show()
visualize_basic_concepts(df)

# --- Data Features for Market Understanding ---
def compare_vs_industry():
    peers = ['TCS.NS', 'INFY.NS', 'WIPRO.NS']
    print("\nðŸ“ˆ Compare RELIANCE with Industry Peers (1Y Return):")
    for peer in peers:
        data = yf.Ticker(peer).history(period='1y')
        if len(data) > 0:
            start_price = data['Close'].iloc[0]
            end_price = data['Close'].iloc[-1]
            change = ((end_price - start_price) / start_price) * 100
            print(f"- {peer}: {change:.2f}%")
        else:
            print(f"- {peer}: Data not available")

# --- Compare vs Industry ---
def compare_vs_industry():
    peers = ['TCS.NS', 'INFY.NS', 'WIPRO.NS']
    print("\nðŸ“ˆ Compare RELIANCE with Industry Peers (1Y Return):")
    for peer in peers:
        data = yf.Ticker(peer).history(period='1y')
        if not data.empty:
            start_price = data['Close'].iloc[0]
            end_price = data['Close'].iloc[-1]
            change = ((end_price - start_price) / start_price) * 100
            print(f"- {peer}: {change:.2f}%")
        else:
            print(f"- {peer}: Data not available")
# --- Backtesting Engine ---
def backtest(df):
    buy = df['Buy_Signal'].dropna()
    sell = df['Sell_Signal'].dropna()
    trades = min(len(buy), len(sell))
    returns = sell.iloc[:trades].values - buy.iloc[:trades].values
    total_return = np.sum(returns)
    print(f"\nðŸ” Backtesting Summary:\n- Trades: {trades}\n- Total â‚¹ Returns: {total_return:.2f}")
backtest(df)

# --- Sentiment & News Analysis Enhancements ---
print("\nðŸ§  Enhanced Sentiment Insight:")
if sentiment_scores['positive'] > sentiment_scores['negative']:
    print("- Market sentiment leans POSITIVE. Stay informed but optimistic.")
elif sentiment_scores['negative'] > sentiment_scores['positive']:
    print("- Market sentiment is NEGATIVE. Caution advised.")
else:
    print("- Market sentiment is NEUTRAL. Watch for signals.")

# --- News Category Tags ---
def tag_news(news_list):
    categories = {"Earnings": [], "Growth": [], "Partnerships": [], "Regulatory": []}
    for news in news_list:
        l = news.lower()
        if "profit" in l or "revenue" in l:
            categories["Earnings"].append(news)
        if "growth" in l or "expansion" in l:
            categories["Growth"].append(news)
        if "partnership" in l or "deal" in l:
            categories["Partnerships"].append(news)
        if "fine" in l or "regulation" in l:
            categories["Regulatory"].append(news)
    print("\nðŸ—‚ï¸ News Categorized:")
    for cat, items in categories.items():
        print(f"- {cat}: {len(items)} articles")
tag_news(news_data)

# --- Smart Alerts & Tips ---
def smart_alerts_tips(df):
    last = df.iloc[-1]
    print("\nðŸ“¡ Smart Alert:")
    if last['20EMA'] > last['50EMA']:
        print("- EMA Crossover: Bullish trend potential.")
    elif last['20EMA'] < last['50EMA']:
        print("- EMA Crossover: Bearish trend potential.")
    else:
        print("- EMAs are flat: Consolidation possible.")
smart_alerts_tips(df)

# --- Beginner-Friendly Alerts ---
print("\nðŸ”° Beginner Tip: Use signals as hintsâ€”not guarantees. Combine with market trends and news.")

# --- Tools to Explore Safely ---
print("\nðŸ§ª Tools to Explore Safely:")
print("- Paper Trading with TradingView")
print("- Investopedia Simulator")
print("- Zerodha Varsity for education")

# --- Portfolio Risk Meter ---
def show_risk_meter(df1):
    if len(df1) == 0:
        print("\nâš ï¸ Risk Meter: No trades to evaluate risk.")
        return
    wins = df1[df1['Profit or Loss'] == 1]
    win_rate = len(wins) / len(df1)
    if win_rate > 0.6:
        risk = "Low"
    elif win_rate > 0.4:
        risk = "Moderate"
    else:
        risk = "High"
    print(f"\nðŸ“Š Portfolio Risk Meter:\n- Win Rate: {win_rate*100:.1f}%\n- Risk Level: {risk}")
show_risk_meter(df1)


# --- Modified Trade Function with Stop-Loss and Take-Profit ---
def trade_call_with_risk_management(df, stop_loss_pct=0.05, take_profit_pct=0.10):
    Buy_price = []
    Sell_price = []
    stop_loss_price = []
    take_profit_price = []
    signal = 0

    for i in range(len(df)):
        if df['20EMA'].iloc[i] > df['50EMA'].iloc[i]:  # Bullish Signal
            if signal != 1:
                Buy_price.append(df['Close'].iloc[i])
                stop_loss_price.append(df['Close'].iloc[i] * (1 - stop_loss_pct))  # Calculate stop loss
                take_profit_price.append(df['Close'].iloc[i] * (1 + take_profit_pct))  # Calculate take profit
                Sell_price.append(np.nan)
                signal = 1
            else:
                Buy_price.append(np.nan)
                Sell_price.append(np.nan)
                stop_loss_price.append(np.nan)
                take_profit_price.append(np.nan)
        elif df['20EMA'].iloc[i] < df['50EMA'].iloc[i]:  # Bearish Signal
            if signal != -1:
                Buy_price.append(np.nan)
                Sell_price.append(df['Close'].iloc[i])
                stop_loss_price.append(np.nan)
                take_profit_price.append(np.nan)
                signal = -1
            else:
                Buy_price.append(np.nan)
                Sell_price.append(np.nan)
                stop_loss_price.append(np.nan)
                take_profit_price.append(np.nan)
        else:
            Buy_price.append(np.nan)
            Sell_price.append(np.nan)
            stop_loss_price.append(np.nan)
            take_profit_price.append(np.nan)

    return Buy_price, Sell_price, stop_loss_price, take_profit_price

df['Buy_Signal'], df['Sell_Signal'], df['Stop_Loss'], df['Take_Profit'] = trade_call_with_risk_management(df)

# --- Visualize Stock with Risk Management (Stop-Loss, Take-Profit) ---
df_plot = df.tail(200)
addplots = [
    mpf.make_addplot(df_plot['20EMA'], color='blue'),
    mpf.make_addplot(df_plot['50EMA'], color='black'),
    mpf.make_addplot(df_plot['Buy_Signal'], type='scatter', markersize=100, marker='^', color='green'),
    mpf.make_addplot(df_plot['Sell_Signal'], type='scatter', markersize=100, marker='v', color='red'),
    mpf.make_addplot(df_plot['Stop_Loss'], type='scatter', markersize=50, marker='x', color='purple'),
    mpf.make_addplot(df_plot['Take_Profit'], type='scatter', markersize=50, marker='*', color='orange')
]
mpf.plot(df_plot, addplot=addplots, type='candle', style='charles', volume=True, figscale=1.2)

# --- Portfolio Diversification Function ---
def portfolio_diversification(df, tickers=['RELIANCE.NS', 'TCS.NS', 'INFY.NS']):
    diversified_df = pd.DataFrame()

    for ticker in tickers:
        ticker_data = yf.Ticker(ticker).history(period='600d', interval='1d')
        ticker_data['20EMA'] = ticker_data['Close'].ewm(span=20, adjust=False).mean()
        ticker_data['50EMA'] = ticker_data['Close'].ewm(span=50, adjust=False).mean()
        ticker_data['Buy_Signal'], ticker_data['Sell_Signal'], ticker_data['Stop_Loss'], ticker_data['Take_Profit'] = trade_call_with_risk_management(ticker_data)
        diversified_df[ticker] = ticker_data['Close']

    # Visualizing diversified portfolio
    diversified_df.plot(figsize=(10, 5))
    plt.title('Diversified Portfolio Performance')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend(tickers)
    plt.show()

# Visualize the Diversified Portfolio for selected stocks
portfolio_diversification(df, tickers=['RELIANCE.NS', 'TCS.NS', 'INFY.NS', 'WIPRO.NS'])

```
